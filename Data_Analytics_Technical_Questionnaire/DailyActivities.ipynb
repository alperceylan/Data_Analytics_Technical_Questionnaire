```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("DailyActivities.xlsx")
df.rename(columns ={'Area of Interest':'AreaOfInterest'}, inplace=True)
df
```

```python
plt.figure(dpi=120)

people = ["Charles", "Henry", "Susan"]

y_values = df.AreaOfInterest
x1_values = df.Charles
x2_values = df.Henry
x3_values = df.Susan

plt.scatter(x1_values, y_values, marker="o")
plt.scatter(x2_values, y_values, marker="^")
plt.scatter(x3_values, y_values, marker="s")

plt.legend(people, loc="lower center", bbox_to_anchor=(0.5, 1.06), ncol=3)

plt.title("How many hours spend in daily activities? (Total 24 hours)", color="r")
plt.ylabel("(Activities)", color="r")
plt.xlabel("(Hours)", color="r")

plt.xticks(np.arange(min(df.index), max(df.index)+5, 1))
plt.minorticks_on()
plt.grid(b=True, which='major', axis='both', linestyle='--', color='g', alpha=0.3)

plt.savefig("How-many-hours-spend-in-daily-activities.PNG", bbox_inches ="tight", dpi=150)

plt.show()
```
