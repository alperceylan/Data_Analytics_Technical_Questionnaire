```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("country_vaccination_stats_1.csv")
df.head()
```

```python
date = df.date.str.split('/', expand=True)
date.tail()
```

```python
df['year'] = date[2]
df['month'] = date[0]
df['day'] = date[1]
df.year = pd.to_numeric(df.year)
df.month = pd.to_numeric(df.month)
df.day = pd.to_numeric(df.day)
df.date = pd.to_datetime(df.date)
df.tail()
```

```python
pivoted_country = df.pivot(index='date', columns='country', values='daily_vaccinations')
pivoted_country.head()
```

```python
pivoted_country["Argentina"].median()
```

```python
pivoted_country.fillna(pivoted_country.median(), inplace=True)
pivoted_country.head()
```

```python
pivoted_country.fillna(0, inplace=True)
```
